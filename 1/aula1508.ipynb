```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dias = np.arange(1,31)
vendas = np.random.randint(50,150, size=30)
```

```python
df_vendas = pd.DataFrame({'Dia':dias,'Vendas':vendas})
df_vendas
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_vendas['Dia'],df_vendas['Vendas'],marker='*', linestyle='-',color='c')

plt.title('Acompanamento de Vendas Dirias', fontsize=16)
plt.xlabel('Dia do Ms', fontsize=14)
plt.ylabel('Vendas Dirias', fontsize=14)

plt.grid(True)
plt.show()
```

```python
categorias = ['Eletrnicos', 'Roupas', 'Alimentos', 'Mveis', 'Livros']
vendas_totais = [1200,850,950,500,750]
```

```python
df_vendas_categorias = pd.DataFrame({'Categoria':categorias,'Vendas':vendas_totais})
df_vendas_categorias
```

```python
plt.figure(figsize=(10,6))
plt.bar(df_vendas_categorias['Categoria'],df_vendas_categorias['Vendas'], color='darkblue')

plt.title('Vendas Totais Por Categoria De Produto - Ms Atual', fontsize=16)
plt.xlabel('Categoria de Produto', fontsize=14)
plt.ylabel('Vendas Totais', fontsize=14)
```

```python
marcas = ['Apple', 'Samsung', 'Xiaomi', 'OnePlus','Outras']
participacao_mercado = [35,30,15,10,10]
```

```python
df_participacao = pd.DataFrame({'Marca':marcas,'Participao':participacao_mercado})
df_participacao
```

```python
plt.figure(figsize=(8,8))
plt.pie(df_participacao['Participao'],labels=df_participacao['Marca'],autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','orange','lightcoral'])

plt.title('Participao de Mercado de Smartphones',fontsize=16)
plt.show()
```
